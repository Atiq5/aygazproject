```python
from google.colab import files
import pandas as pd
import seaborn as sns

# Set Seaborn figure size
sns.set(rc={'figure.figsize': (5, 5)})

# Upload the 'airbnb.csv' file
uploaded = files.upload()

# Get the file name from the uploaded keys
file_name = list(uploaded.keys())[0]

try:
    # Load the data using pandas
    df_airbnb = pd.read_csv(file_name)

    # Display the first five rows of the DataFrame
    print(df_airbnb.head(5))

except FileNotFoundError:
    print(f"Error: The file '{file_name}' was not found in the /content/ directory.")
except Exception as e:
    print(f"An error occurred while loading the file: {str(e)}")
```
